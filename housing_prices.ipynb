```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.tree import DecisionTreeRegressor
import pickle 
```

```python
data = pd.read_csv('/home/armen/Downloads/train.csv')

data = data[["full_sq", "sub_area", "floor", "product_type", "price_doc"]].dropna()
```

```python
X = data[["full_sq", "sub_area", "floor", "product_type"]]
y = data["price_doc"]
X
```

```python
le = LabelEncoder()
le.fit(X["sub_area"])
area_list = le.classes_
X["sub_area"] = le.transform(X["sub_area"])
le.fit(X["product_type"])
type_list = le.classes_
X["product_type"] = le.transform(X["product_type"])
X
```

```python
area_dict = dict(zip(area_list, list(range(len(area_list)))))
type_dict = dict(zip(type_list, list(range(len(type_list)))))
```

```python
area_dict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
dt = DecisionTreeRegressor()
parameters = {'max_depth':[None, 2, 3, 4, 5, 6, 7, 8], 'min_samples_split':[2, 3, 4, 5, 6, 7, 8, 9, 12], 'random_state':[42],
             'max_features':('auto', 'sqrt', 'log2', None)}
reg = GridSearchCV(dt, parameters, cv=3)
reg.fit(X_train, y_train)
print(reg.best_params_)
print(reg.best_score_)
best_est = reg.best_estimator_
```
